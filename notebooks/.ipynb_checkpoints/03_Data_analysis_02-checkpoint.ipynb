```python
import psycopg2
import pandas as pd
import matplotlib.pyplot as plt 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from psycopg2 import connect
from sqlalchemy import create_engine
```

 ## Connection to the database

```python
username = 'postgres'
password = 'password'

host = 'localhost'
database = 'airlines'
port = 5432
```

```python
url = f"postgresql://{username}:{password}@{host}:{port}/{database}"
engine = create_engine(url)
```

```python
def read_sql_table(table_name):
    msg = f'Pobieram dane z tabeli: {table_name}'
    print(msg)
    
    file = pd.read_sql("SELECT * FROM" +" " + table_name +';'"", engine)
    return file
```

```python
filename = "../data/processed/flight_df_01.csv"
flight_df = pd.read_csv(
        filename,
        sep=';',
        decimal='.',
        encoding='UTF-8',
    )

flight_df
```

 # Enriching with `aircraft`

```python
aircraft_df_first = read_sql_table('aircraft')

aircraft_df = aircraft_df_first.copy()
aircraft_df
```

```python
aircraft_df = aircraft_df.drop(['id', 'number_of_seats'], axis = 1)
```

```python
aircraft_df = aircraft_df.drop_duplicates()
```

```python
aircraft_df_is_duplicated = aircraft_df.duplicated(subset='tail_num')
aircraft_df_duplicated = aircraft_df.loc[aircraft_df_is_duplicated]
aircraft_df_duplicated
```

 ## Modification `aircraft_df`

```python
aircraft_df = (aircraft_df
               .groupby(by='tail_num', as_index=False)['manufacture_year']
               .max()
              )
aircraft_df
```

 ## Join `aircraft_df` and `flight_df`

```python
tmp_flight_df = pd.merge(
    left=flight_df,
    right=aircraft_df,
    how='left',
    on=['tail_num']
)

tmp_flight_df.info()
```

```python
flight_df.shape[0] == tmp_flight_df.shape[0]
```

```python
flight_df = tmp_flight_df.copy()
```

 ## Delay vs. year of production of the aircraft part 1

```python
delays_by_manufacture_year_df = (flight_df
     .groupby(by=['manufacture_year'], as_index=False)
     .agg({'is_delayed':['mean', 'size']})
    )
delays_by_manufacture_year_df
```

```python
plt.figure(figsize=(15,10))

plt.scatter(x=delays_by_manufacture_year_df['manufacture_year'],
            y=delays_by_manufacture_year_df[('is_delayed', 'mean')],
            c=delays_by_manufacture_year_df[('is_delayed', 'size')]
            )

plt.title('Manufacture_year vs. Dep_delay')

plt.xlabel('Manufacture_year')
plt.ylabel('Dep_delay')
```

```python
delays_by_manufacture_year_df_over_10000 = delays_by_manufacture_year_df[
    ('is_delayed', 'size')] >10000
to_plot = delays_by_manufacture_year_df.loc[delays_by_manufacture_year_df_over_10000]

plt.figure(figsize=(15,10))

plt.scatter(
    x=to_plot['manufacture_year'],
    y=to_plot[('is_delayed', 'mean')],
    c=to_plot[('is_delayed', 'size')]
)

plt.title('Manufacture_year vs. Dep_delay')

plt.xlabel('manufacture_year')
plt.ylabel('dep_delay')
```

 ## Delay vs. year of production of the aircraft part 2

```python
bins = range(1990, 2021, 3)
flight_df['manufacture_year_agg'] = pd.cut(flight_df['manufacture_year'], bins)
flight_df
```

```python
flight_delays_by_manufacture_year_agg_df = (flight_df
                                     .groupby(by='manufacture_year_agg')['is_delayed']
                                     .agg(['mean', 'size'])
                                     )
flight_delays_by_manufacture_year_agg_df
```

```python
plt.figure(figsize=(15,10))

plt.scatter(
    x=flight_delays_by_manufacture_year_agg_df.index.astype(str),
    y=flight_delays_by_manufacture_year_agg_df['mean'],
    c=flight_delays_by_manufacture_year_agg_df['size']
)

plt.title('Delay %')
plt.xticks(rotation=45)

plt.xlabel('Year')
plt.ylabel('Delay')
```

```python
top_manufactured_df_2 = (delays_by_manufacture_year_df
                         .sort_values(by=[('is_delayed', 'size')], ascending=False)
                         .head(5)
                        )
top_manufactured_df_2                       
```
